# 흑백처리

```python
image_bgr = cv2.imread(fname)
```

```python

if image_bgr is None:
    print(f'Error: 파일을 불러 올 수 없음. {car_image_path}')
else:
    # 1. BGR 이미지를 RGB로 변환 (Matplotlib 출력을 위함)
    image_rgb_ori = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    # 2. 흑백(Grayscale) 이미지로 변환
    image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
    print(f'흑백 이미지 형태 : {image_gray.shape}')

    # 3. 흑백 이미지를 BGR 값(3채널)을 가진 이미지로 변환
    # Matplotlib은 RGB를 기대하므로, 다시 RGB로 변환합니다.
    image_rgb_from_gray = cv2.cvtColor(image_gray, cv2.COLOR_GRAY2RGB)
    print(f'RGB 변환 후 이미지 형태 : {image_rgb_from_gray.shape}')

    # 4. 1행 3열 레이아웃 설정 (figsize를 넉넉하게 조정)
    plt.figure(figsize=(15, 5)) # 가로 크기를 15로 늘려서 3개 이미지를 나란히 배치

    # 첫 번째 이미지: 원본 (RGB)
    plt.subplot(1, 3, 1) # 1행 3열 중 1번째
    plt.imshow(image_rgb_ori) # Matplotlib 출력에 맞춘 RGB 원본
    plt.title('1. Original (RGB)')
    plt.axis('off')

    # 두 번째 이미지: 일반적인 흑백 출력 (Grayscale, cmap='gray' 사용)
    plt.subplot(1, 3, 2) # 1행 3열 중 2번째
    plt.imshow(image_gray, cmap='gray')
    plt.title('2. Grayscale (1 Channel)')
    plt.axis('off')

    # 세 번째 이미지: 흑백을 RGB 3채널로 변환한 이미지
    plt.subplot(1, 3, 3) # 1행 3열 중 3번째
    plt.imshow(image_rgb_from_gray) # 3채널이지만 모두 같은 값이라 흑백으로 보임
    plt.title('3. Grayscale to RGB (3 Channels)')
    plt.axis('off')

    plt.tight_layout() # 이미지와 제목이 겹치지 않게 레이아웃 조정
    plt.show()
```

# 누끼

## Gray 1 Channel

```python
img = Image.fromarray(image_gray).convert("RGB")  # numpy → PIL 변환
out = remove(img)
out_path = "image_gray_no_bg.png"
out.save(out_path)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(img); plt.axis('off'); plt.title('Original')
plt.subplot(1,2,2); plt.imshow(out); plt.axis('off'); plt.title('Removed BG')
plt.show()
```
