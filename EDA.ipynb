```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# 1ï¸âƒ£ ì—‘ì…€ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
file_path = "/content/Final.xlsx"
df = pd.read_excel(file_path)

# 2ï¸âƒ£ ì£¼í–‰ê±°ë¦¬ ì»¬ëŸ¼ì—ì„œ ê²°ì¸¡ì¹˜ ì œê±°
mileage = df['mileage'].dropna()

# 3ï¸âƒ£ 50,000 ë‹¨ìœ„ êµ¬ê°„ ì„¤ì •
bins = np.arange(0, 200000 + 50000, 50000)
labels = [
    "0 ~ 50,000 km",
    "50,000 ~ 100,000 km",
    "100,000 ~ 150,000 km",
    "150,000 ~ 200,000 km",
]
extra_label = "200,000 km ì´ìƒ"

# 4ï¸âƒ£ ì£¼í–‰ê±°ë¦¬ êµ¬ê°„í™”
mileage_bins = pd.cut(mileage, bins=bins, labels=labels, right=True)
mileage_bins = mileage_bins.astype('object')
mileage_bins[mileage > 200000] = extra_label

# 5ï¸âƒ£ êµ¬ê°„ë³„ ì¹´ìš´íŠ¸ ê³„ì‚°
mileage_counts = mileage_bins.value_counts().sort_index()
desired_order = labels + [extra_label]
mileage_counts = mileage_counts.reindex([x for x in desired_order if x in mileage_counts.index])

# 6ï¸âƒ£ ë¹„ìœ¨ ê³„ì‚°
ratios = mileage_counts / mileage_counts.sum() * 100

# 7ï¸âƒ£ ì»¤ìŠ¤í…€ ì»¬ëŸ¬ë§µ ì •ì˜ (íšŒìƒ‰â†’ë‚¨ìƒ‰â†’ì£¼í™©â†’ë…¸ë‘)
colors_list = ["#6e6e6e", "#1f3c88", "#f28e1c", "#ffe66d"]
custom_cmap = mcolors.LinearSegmentedColormap.from_list("gray_blue_orange_yellow", colors_list)
colors = custom_cmap(1 - (ratios / ratios.max()))  # ë¹„ìœ¨ì´ ë†’ì„ìˆ˜ë¡ íšŒìƒ‰ ìª½ìœ¼ë¡œ

# âœ… ë¼ë²¨ êµ¬ì„± ("í•­ëª©ëª…\nxx.x%")
labels_with_pct = [f"{label}\n{ratio:.1f}%" for label, ratio in zip(mileage_counts.index, ratios)]

# 8ï¸âƒ£ íŒŒì´ì°¨íŠ¸ ì‹œê°í™”
plt.figure(figsize=(6, 5))
plt.pie(
    mileage_counts,
    labels=labels_with_pct,
    colors=colors,
    startangle=90,
    counterclock=False,
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 11},
    labeldistance=1.15  # ë¼ë²¨ ìœ„ì¹˜ ì‚´ì§ ë°”ê¹¥ìª½
)

# 9ï¸âƒ£ ê·¸ë˜í”„ ê¾¸ë¯¸ê¸°
plt.axis('equal')  # ì›í˜• ìœ ì§€
plt.tight_layout()
plt.show()

# ğŸ”Ÿ ê²°ê³¼ ì¶œë ¥
print(mileage_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# 1ï¸âƒ£ ì—‘ì…€ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
file_path = "/content/Final.xlsx"
df = pd.read_excel(file_path)

# 2ï¸âƒ£ íŒë§¤ê¸°ê°„ êµ¬ê°„ ì„¤ì • (0~100ì¼: 10ì¼ ë‹¨ìœ„, 100ì¼ ì´ìƒ ë¬¶ìŒ)
bins = list(range(0, 100, 10)) + [100, float('inf')]
labels = [f"{i}~{i+10}ì¼" for i in range(0, 100, 10)] + ["100ì¼ ì´ìƒ"]

def make_profit_summary(df_input, name):
    df_temp = df_input.dropna(subset=['sales_period', 'profit']).copy()
    df_temp['sales_period'] = pd.to_numeric(df_temp['sales_period'], errors='coerce')
    df_temp['profit'] = pd.to_numeric(df_temp['profit'], errors='coerce')
    df_temp['íŒë§¤ê¸°ê°„êµ¬ê°„'] = pd.cut(df_temp['sales_period'], bins=bins, labels=labels, right=False, include_lowest=True)

    profit_summary = (
        df_temp.groupby('íŒë§¤ê¸°ê°„êµ¬ê°„', observed=True)['profit']
        .mean()
        .dropna()
        .apply(lambda x: round(x, -4))
    )
    profit_summary.name = name
    return profit_summary

# 3ï¸âƒ£ ì „ì²´ / ê·¸ëœì € HG / ìŠ¤íŒŒí¬ ê°ê° ìš”ì•½
overall = make_profit_summary(df, "ì „ì²´")
grandeur = make_profit_summary(df[df['model'].astype(str).str.contains("ê·¸ëœì € HG", na=False)], "ê·¸ëœì € HG")
spark = make_profit_summary(df[df['model'].astype(str).str.contains("ìŠ¤íŒŒí¬", na=False)], "ìŠ¤íŒŒí¬")

# 4ï¸âƒ£ í†µí•©
profit_df = pd.concat([overall, grandeur, spark], axis=1)

# 5ï¸âƒ£ ì‹œê°í™”
plt.figure(figsize=(12, 7))
colors = {'ì „ì²´': 'black', 'ê·¸ëœì € HG': 'royalblue', 'ìŠ¤íŒŒí¬': 'tomato'}

for col in profit_df.columns:
    plt.plot(profit_df.index, profit_df[col], marker='o', linewidth=2, color=colors[col], label=col)

    # ê° ì  ìœ„ì— ë¼ë²¨ í‘œì‹œ
    for x, y in zip(profit_df.index, profit_df[col]):
        if pd.notna(y):
            plt.text(x, y + (profit_df.max().max() * 0.015), f"â‚©{y:,.0f}",
                     ha='center', va='bottom', fontsize=8, color=colors[col])

# 6ï¸âƒ£ yì¶• ì„¤ì •
ax = plt.gca()
ax.set_ylim(-1_000_000, 1_400_000)
ax.set_yticks(np.arange(-1_000_000, 1_400_001, 200_000))

# â‚© í¬ë§· (ìŒìˆ˜ì— -â‚© í‘œì‹œ)
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"-â‚©{abs(x):,.0f}" if x < 0 else f"â‚©{x:,.0f}")
)

# 7ï¸âƒ£ ìŠ¤íƒ€ì¼ ì„¸ë¶€ ì„¤ì •
plt.xticks(fontsize=13)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="ì°¨ì¢…", fontsize=10)

plt.tight_layout()
plt.show()
```
